```python
import pandas as pd

appartment = pd.read_csv('apartment_links.csv')
```

```python
appartment
```

```python
house = pd.read_csv('house_links.csv')
house
```

```python
appartment_house_link = pd.concat([appartment, house])
appartment_house_link.drop(['Unnamed: 0'], axis=1,inplace=True)
```

```python
appartment_house_link.reset_index(inplace=True)
```

```python
appartment_house_link.drop(['index'], axis=1,inplace=True)
```

```python
appartment_house_link
```

```python
appartment_house_link.to_csv('appartment_house_links.csv',index=False)
```
