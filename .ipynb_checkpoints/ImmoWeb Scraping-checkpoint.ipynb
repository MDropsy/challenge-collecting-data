```python
import json
import requests
from bs4 import BeautifulSoup
import numpy as np

pages = np.arange(1, 334)

urls=[]

for page in pages:
    
    url = 'https://www.immoweb.be/en/search/house-and-apartment/for-sale?countries=BE&page='+str(page)+'&orderBy=relevance'
  
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')

    data = json.loads( soup.find('iw-search')[':results'] )

    for d in data:

        url = 'https://www.immoweb.be/en/classified/{}'.format(d['id'])

        urls.append(url)
```

```python
import requests
import lxml.html as lh
import pandas as pd

url="https://www.immoweb.be/en/classified/8900838"

#Create a handle, page, to handle the contents of the website
page = requests.get(url)

#Store the contents of the website under doc
doc = lh.fromstring(page.content)

#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')

el=[]

for tr in tr_elements:
    
    name=tr.text_content()
    el.append(name)

df = {}

for item in el:

    items = list(filter(None, map(str.strip, item.split("\n"))))
    
    try:
        df[items[0]] = items[1]
        
    except IndexError:
    
        
        continue    
```
